```python
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
import pandas as pd
d= sns.load_dataset("titanic")
d
```

```python
d.isnull().sum()
```

```python
d.dropna(subset = ['embarked'  ,'embark_town'] , inplace= True)
```

```python
d.isnull().sum()
```

```python
d.duplicated().sum()
```

```python
d.drop_duplicates(inplace = True)
```

```python
d.duplicated().sum()
d
```

```python
d["age"].groupby([d["who"], d["sex"]]).mean()
```

```python
d["age"] = d["age"].groupby([d["who"], d["sex"]]).transform(lambda x : x.fillna(x.mean()))
d
```

```python
fig = plt.figure()
sns.countplot(x = 'who' , hue= 'alive',data= d  )
```
